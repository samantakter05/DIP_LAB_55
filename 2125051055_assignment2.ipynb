```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

```python
image = '/content/drive/MyDrive/DIP.jpg'
```

```python
image = cv2.imread(image)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
resized = cv2.resize(image, (256, 256))
```

```python
rotated = cv2.rotate(resized, cv2.ROTATE_180)
```

```python
flipped = cv2.flip(rotated, 0)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(10, 5))

axes[0].imshow(resized)
axes[0].set_title("Resized (256x256)")
axes[0].axis("off")

axes[1].imshow(rotated)
axes[1].set_title("Rotated 180")
axes[1].axis("off")

axes[2].imshow(flipped)
axes[2].set_title("Vertical Flip")
axes[2].axis("off")


plt.show()
```
