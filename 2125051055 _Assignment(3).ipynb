Create two 400x400 images with black background : The first image should contain a green rectangle at the center. The second image should contain a magenta circle at the center.

Perform the following operations using OpenCV and display the results:

Bitwise AND (cv2.bitwise_and): Show the overlapping region of the two images.

Bitwise OR (cv2.bitwise_or): Combine both shapes while retaining their colors.

Bitwise XOR (cv2.bitwise_xor): Retains only the non-overlapping parts of both images.

Bitwise NOT (cv2.bitwise_not): Inverts the colors of each shape separately.

Element-wise Multiplication (cv2.multiply): Observe the effect on overlapping areas.

Alpha Blending (cv2.addWeighted): Blend both images with transparency effects.

```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image1 = np.zeros((400, 400, 3), dtype=np.uint8)

cv2.rectangle(image1, (100, 100),( 300, 300),(0, 255, 0), -1)
cv2_imshow(image1)
```

```python
image2 = np.zeros((400, 400, 3), dtype=np.uint8)
cv2.circle(image2, (200, 200), 100, (255, 0, 255), -1)
cv2_imshow(image2)
```

```python
bitwise_and = cv2.bitwise_and(image1, image2)
cv2_imshow(bitwise_and)
```

```python
bitwise_or = cv2.bitwise_or(image1, image2)
cv2_imshow(bitwise_or)
```

```python
bitwise_not1 = cv2.bitwise_not(image1)
cv2_imshow(bitwise_not1)
```

```python
bitwise_not2 = cv2.bitwise_not(image2)
cv2_imshow(bitwise_not2)
```

```python
elementwise_multiplication = cv2.multiply(image1, image2)
cv2_imshow(elementwise_multiplication)
```

```python
alpha = 0.5
alpha_blending_result = cv2.addWeighted(image1, alpha, image2, 1 - alpha, 0)
cv2_imshow(alpha_blending_result)
```
